```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
hospital_df = pd.read_csv('AYUSHHospitals.csv')
hospital_df
```

```python
state_cnt = hospital_df['State / UT'].value_counts()

plt.style.use('ggplot')
state_cnt.plot(kind='bar')    ##bar plot shows  categorical data whereas ##histogram presents numerical data
plt.show()
```

```python
hospitalBeds_df = pd.read_csv('Hospitals and Beds.csv')
hospitalBeds_df
```

```python
Name_of_state_cnt = hospitalBeds_df['Name of State'].value_counts()
plt.style.use('ggplot')
Name_of_state_cnt.plot(kind='bar')
plt.show()
```
